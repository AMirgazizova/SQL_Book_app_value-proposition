- **Find the number of books released after January 1, 2000**

```python
query = '''SELECT
COUNT(*) AS cnt,
COUNT(book_id) AS books_id_cnt,
COUNT(DISTINCT book_id) AS books_amount_uniq
FROM
books
WHERE
CAST(books.publication_date AS date) >= '2000-01-01'
'''

pd.io.sql.read_sql(query, con = engine)
```

<font color = 'purple'>**The number of books released after January 1, 2000 is 821**</font>

- **Find the number of user reviews and the average rating for each book**

```python
query = '''SELECT
COUNT(rating_id) AS review_amount,
COUNT(DISTINCT rating_id) review_amount_uniq
FROM
ratings
'''

pd.io.sql.read_sql(query, con = engine)
```

```python
query = '''SELECT
books.book_id AS id,
books.title AS title,
COUNT(DISTINCT reviews.review_id) AS reviews_count,
AVG(ratings.rating) AS avg_rating
FROM books
RIGHT JOIN reviews ON reviews.book_id = books.book_id
RIGHT JOIN ratings ON ratings.book_id = books.book_id
GROUP BY
id, title
ORDER BY reviews_count DESC;'''

pd.io.sql.read_sql(query, con = engine)
```

<font color = 'purple'>**The top books by the average rating are Emily of New Moon and Stolen Hard Times and by number of user reviews Twilight  and The Road**</font>

- **Identify the publisher that has released the greatest number of books with more than 50 pages (this will help you exclude brochures and similar publications from your analysis)**

```python
query = '''SELECT
publishers.publisher AS publisher,
COUNT(DISTINCT books.book_id) AS max_books
FROM books
INNER JOIN publishers ON publishers.publisher_id = books.publisher_id
WHERE
    num_pages >= 50
GROUP BY
publisher
ORDER BY max_books DESC;'''

pd.io.sql.read_sql(query, con = engine)
```

<font color = 'purple'>**The publisher that has released the greatest number of books with more than 50 pages is Penguin Books with 42 books**</font>

- **Identify the author with the highest average book rating: look only at books with at least 50 ratings**

```python
query = '''SELECT
books.book_id AS books_id,
books.title AS title,
authors.author AS author,
AVG(book_rating) AS avg_book_rating

FROM 
books

LEFT JOIN authors ON books.author_id = authors.author_id

INNER JOIN( SELECT 
book_id, 
AVG(rating)
FROM 
ratings 
GROUP BY book_id
HAVING COUNT(rating)  >=  50) AS book_rating ON books.book_id = book_rating.book_id

GROUP BY
books_id, title, author, avg_book_rating
ORDER BY avg_book_rating DESC
'''

pd.io.sql.read_sql(query, con = engine)
```

<font color = 'purple'>**The author with the highest average book rating is J.K. Rowling/Mary GrandPré - Harry Potter and the Prisoner of Azkaban has the highest avg book rating 4.4**</font>

- **Find the average number of text reviews among users who rated more than 50 book**

```python
book_id = f'''SELECT
AVG(SUBQUERY.n_reviews) AS avg_text_rev
FROM

   (SELECT 
    username,
    COUNT(DISTINCT review_id) AS n_reviews
    FROM reviews
    WHERE username IN
       (SELECT username AS username
        FROM ratings
        GROUP BY username
        HAVING COUNT(DISTINCT book_id) >50)
        GROUP BY username) AS SUBQUERY;'''



pd.io.sql.read_sql(book_id , con = engine)
```
